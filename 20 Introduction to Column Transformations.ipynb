### withColumn() method

In this section, we demonstrate how to add a new column using the `withColumn()` method. The new column `population_m` represents the population in millions.

```python
from pyspark.sql.functions import col

# Add a new column 'population_m' by dividing 'population' by 1,000,000
df.withColumn("population_m", col("population")/1000000).show(5)
```

Since PySpark DataFrames are immutable, the `withColumn()` method returns a new DataFrame with the transformation applied. Here, we assign the transformed DataFrame to a new variable `df_2` to retain the changes.

```python
# Reassign the transformed DataFrame to a new variable
df_2 = df.withColumn("population_m", col("population")/1000000)

# Display the original DataFrame (df) and the transformed DataFrame (df_2)
df.show(5)
df_2.show(5)
```

We can also transform an existing column. In this example, we convert the `country_name` column to uppercase using the `upper()` function.

```python
from pyspark.sql.functions import upper

# Convert the 'country_name' column to uppercase
df.withColumn("country_name", upper("country_name")).show(5)
```

### Chaining Multiple Transformations with withColumn()

Here, we chain multiple `withColumn()` calls to apply several transformations in sequence. Each call creates a new DataFrame, which can be less efficient for many transformations.

```python
df.withColumn("country_name", upper("country_name"))\
  .withColumn("population_m", col("population")/1000000)\
  .show(5)
```

### Applying Multiple Transformations with select()

Using the `select()` method, we can apply multiple transformations in a single step, which is generally more efficient than chaining multiple `withColumn()` calls. In this example:
- We convert `country_name` to uppercase and alias it as `country_name`.
- We retain the `continent` column.
- We create a new column `population_m` by dividing `population` by 1,000,000.
- We retain the `area_km2` column.

```python
df.select(
    upper("country_name").alias("country_name"),  # Convert country names to uppercase
    "continent",                                  # Retain the 'continent' column
    (col("population")/1000000).alias("population_m"),  # Create 'population_m' column
    "area_km2"                                    # Retain the 'area_km2' column
).show(5)
```
